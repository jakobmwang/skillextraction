```python
# dependencies
import os
import re
import numpy as np
import pandas as pd
from google.colab import drive, userdata

# file management
drive.mount('/content/drive')
WORK_DIR = '/content/drive/MyDrive/Projects/skillextraction'

# work dir shortcut function
def work_dir(*args):
    return os.path.join(WORK_DIR, *args)
```

```python
# get esco
esco_en = pd.read_csv(work_dir('ESCO', 'ESCO dataset - v1.1.2 - classification - en - csv', 'skills_en.csv'))
```

```python
# get real sentences
reals = pd.concat([pd.read_csv(work_dir('Annotated_data', s))
                   for s in os.listdir(work_dir('Annotated_data')) if re.match(r'^sentences\_[0-9]+\.csv$', s)])

# divide
skills = reals[reals['conceptUri'].notna() & (reals['conceptUri'] != '') & (reals['is_skill'] == 1)].tail(100)
nonskills = reals[(reals['conceptUri'].isna() | (reals['conceptUri'] == '')) & (reals['is_skill'] == 0)].tail(100)

# map label
skills['label'] = skills['conceptUri'].map(dict(zip(esco_en['conceptUri'], esco_en['preferredLabel'])))

# readable sentences
pd.set_option('display.max_colwidth', None)

# check
skills
```
